```
import pandas as pd
import tensorflow as tf 
from tensorflow import keras 
from tensorflow.keras import Sequential 
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
import numpy as np

bd = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/bank_note_data/training_set_label.csv" )
```

```
X = bd.drop('Class',axis=1) #input variable
y = bd.Class # output variable
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=40)
```

```
model = Sequential() # define a squential model
model.add(Dense(32, activation='relu', input_shape=(X_train.shape[1],)))   #contains 32 perceptrons in 1 layer with relu activator
model.add(Dense(16, activation='relu'))  #contains 16 perceptrons in 2 layer with relu activator
model.add(Dense(8, activation='relu'))  #contains 8 perceptrons in 3 layer with relu activator
model.add(Dense(1, activation='sigmoid'))  #contains 1 for output perceptron in last layer with activator sigmoid for better performace we have used sofmax but it was little better than that though no difference
```

```
from tensorflow.keras.optimizers import RMSprop
optimizer = RMSprop(0.001)  # Here, we have set our learning rate as 0.001
model.compile(loss='binary_crossentropy', optimizer = optimizer , metrics=['accuracy'])
```

```
# printing the summary of the model
model.summary()
```

```
history = model.fit(X_train, y_train, validation_split=0.2, epochs=200, batch_size=10, verbose=2) #tried 100,150,200 epoches after 150 it is giving 100% acc
```

```
model.evaluate(X_test, y_test)
```
