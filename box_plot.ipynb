```
import pandas as pd

data = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/liver_patient.csv")
data
```

```
import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure(figsize=(10,5))

#matplot
ax = fig.add_subplot(121)
ax.set_title('Box plot Box plot of Total Protiens Seaborn Matplotlib')
data.Total_Protiens.plot(kind='box', ax=ax)


#sns
ax = fig.add_subplot(122)
ax.set_title('Box plot of Total Protiens Seaborn')
sns.boxplot(data.Total_Protiens)

plt.tight_layout()
plt.show()
```
