```
import pandas as pd
from sklearn.model_selection import train_test_split
sd  = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/startupdata/training_set_label.csv" )
td = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/startupdata/testing_set_label.csv')
sd = sd.drop(columns=['Unnamed: 0','Unnamed: 6','state_code.1','closed_at','object_id','is_CA' ,'is_NY' ,'has_VC' ,'has_roundB' ,'has_roundC' ,'has_roundD' ,'has_roundA' ,'is_ecommerce' ,'is_biotech' ,'is_gamesvideo', 'is_enterprise' ,'is_mobile' ,'is_web', 'is_software', 'is_otherstate', 'has_angel' ,'is_MA', 'is_othercategory' ,'state_code'],axis =1)
td = td.drop(columns=['Unnamed: 0','Unnamed: 6','state_code.1','closed_at','object_id','is_CA' ,'is_NY' ,'has_VC' ,'has_roundB' ,'has_roundC' ,'has_roundD' ,'has_roundA' ,'is_ecommerce' ,'is_biotech' ,'is_gamesvideo', 'is_enterprise' ,'is_mobile' ,'is_web', 'is_software', 'is_otherstate', 'has_angel','is_MA', 'is_othercategory' ,'state_code'],axis =1)
```

```
X = sd.drop('status',axis=1)
sd['age_first_milestone_year'].fillna(sd['age_first_milestone_year'].mode()[0], inplace=True)
sd['age_last_milestone_year'].fillna(sd['age_last_milestone_year'].mode()[0], inplace=True)
td['age_first_milestone_year'].fillna(td['age_first_milestone_year'].mode()[0], inplace=True)
td['age_last_milestone_year'].fillna(td['age_last_milestone_year'].mode()[0], inplace=True)
# Import label encoder 
from sklearn import preprocessing
# label_encoder object knows how to understand word labels. 
le = preprocessing.LabelEncoder()
# Encode labels in column 'Country'. 
X['id'] = le.fit_transform(X['id'])
X['city'] = le.fit_transform(X['city'])
X['zip_code'] = le.fit_transform(X['zip_code'])
X['name'] = le.fit_transform(X['name'])
X['category_code'] = le.fit_transform(X['category_code'])
X['first_funding_at'] = le.fit_transform(X['first_funding_at'])
X['last_funding_at'] = le.fit_transform(X['last_funding_at'])
X['founded_at'] = le.fit_transform(X['founded_at'])
td['id'] = le.fit_transform(td['id'])
td['city'] = le.fit_transform(td['city'])
td['name'] = le.fit_transform(td['name'])
td['zip_code'] = le.fit_transform(td['zip_code'])
td['category_code'] = le.fit_transform(td['category_code'])
td['first_funding_at'] = le.fit_transform(td['first_funding_at'])
td['last_funding_at'] = le.fit_transform(td['last_funding_at'])
td['founded_at'] = le.fit_transform(td['founded_at'])

y = sd[['status']]
```

```
newdata = pd.concat([X, y], axis=1, sort=False)
new_train = pd.DataFrame(newdata) 
new_test = pd.DataFrame(td) 
# saving the dataframe 
new_train.to_csv('hptrain.csv')
new_test.to_csv('hptest.csv')
new_train
```

```
from autogluon import TabularPrediction as task
test_data = task.Dataset(file_path='hptest.csv')
train_data = task.Dataset(file_path="hptrain.csv")
predictor = task.fit(train_data=train_data, label='status')
y_pred = predictor.predict(test_data)
```
